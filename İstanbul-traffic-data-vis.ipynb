```python
import geopandas

df = geopandas.read_file("/kaggle/input/trafficvol/istanbul-districts (1).json")
```

```python
df_crs = df.to_crs("EPSG:2263")
df_crs["area"] = df_crs.area
```

```python
import contextily as ctx

df_area = df_crs.plot("area" , legend=True,cmap="inferno_r",alpha=0.7,figsize=[11,11])
ctx.add_basemap(df_area,crs="EPSG:2263",source=ctx.providers.OpenStreetMap.HOT)
plt.title("sik")
```
